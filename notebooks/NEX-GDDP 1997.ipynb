## Load up a single model's worth of data for 1997

```python
import numpy as np
import pandas as pd
import xray
```

```python
def get_dataset(model):
    ds = xray.open_dataset("data/NEX-GDDP-1997/pr_day_BCSD_historical_r1i1p1_{0}_1997.nc".format(model))
    ds = ds.merge(xray.open_dataset("data/NEX-GDDP-1997/tasmin_day_BCSD_historical_r1i1p1_{0}_1997.nc".format(model)))
    ds = ds.merge(xray.open_dataset("data/NEX-GDDP-1997/tasmax_day_BCSD_historical_r1i1p1_{0}_1997.nc".format(model)))
    return ds
```

```python
albany_wu = pd.read_csv("data/NEX-GDDP-1997/weather_underground_ALB_1997.csv")
albany_wu["EST"] = pd.to_datetime(albany_wu["EST"])
albany_wu.set_index("EST", inplace=True)
```

```python
# Convert to kg m-2 s-1
albany_wu['pr'] = (albany_wu['PrecipitationIn'] * 24.5) / 86400.
```

```python
# Roughly Guilderland to Troy
lat = 42.875

# Roughly Albany to Cliftonpark
lon = 360. -73.625
```

## Model Comparison

```python
models = [
"ACCESS1-0",
"BNU-ESM",
"CCSM4",
"CESM1-BGC",
"CNRM-CM5",
"CSIRO-Mk3-6-0",
"CanESM2",
"GFDL-CM3",
"GFDL-ESM2G",
"GFDL-ESM2M",
"IPSL-CM5A-LR",
"IPSL-CM5A-MR",
"MIROC-ESM-CHEM",
"MIROC-ESM",
"MIROC5",
"MPI-ESM-LR",
"MPI-ESM-MR",
"MRI-CGCM3",
"NorESM1-M",
"bcc-csm1-1",
"inmcm4"]

ds = {}

for m in models:
    dataset = get_dataset(m)
    ds[m] = dataset.loc[dict(lat=lat, lon=lon)]
    del dataset
```

```python
from matplotlib.pylab import plt
fig, ax = plt.subplots(1,1, figsize=(16., 8.))

for m, d in ds.items():
    d.to_dataframe()['pr'].plot(label=m, ax=ax, alpha=0.2, style='o')

albany_wu['pr'].plot(label="KALB", ax=ax, style='r^')
plt.title("NEX GDDP Precipitation Models, historical for 1997 with Weather Underground data for KALB")
plt.legend()
```

### Min/Max Temp

```python
#fig, ax = plt.subplots(1,1, figsize=(16., 8.))

gcms = pd.concat({k:d.to_dataframe()[['tasmin', 'tasmax', 'pr']] for k,d in ds.items()}, axis=1)
gcm_desc = gcms.stack(level=0).groupby(level=0).describe().unstack()
```

```python
fix, ax = plt.subplots(1,1, figsize=(16., 8.))

def to_F(K):
    return ((K * 9.0) / 5.0) - 459.67

gcm_desc['tasmax']['mean'].apply(to_F).plot(ax=ax, label="GDDP Mean Max Temp")
plt.fill_between(gcm_desc['tasmax'].index, 
                 gcm_desc['tasmax']['min'].apply(to_F), 
                 gcm_desc['tasmax']['max'].apply(to_F),
                alpha=0.3)


gcm_desc['tasmin']['mean'].apply(to_F).plot(ax=ax, label="GDDP Mean Min Temp")
plt.fill_between(gcm_desc['tasmin'].index, 
                 gcm_desc['tasmin']['min'].apply(to_F), 
                 gcm_desc['tasmin']['max'].apply(to_F),
                alpha=0.3)

plt.title("Mean Max Temp, Mean Min Temp; shaded region is min/max for 21 GCMs in CMIP5")
plt.legend(loc=4)
```

```python
fix, ax = plt.subplots(1,1, figsize=(16., 8.))

gcm_desc['tasmax']['mean'].apply(to_F).plot(ax=ax, label="GDDP Mean Max Temp")
plt.fill_between(gcm_desc['tasmax'].index, 
                 gcm_desc['tasmax']['min'].apply(to_F), 
                 gcm_desc['tasmax']['max'].apply(to_F),
                alpha=0.1)


gcm_desc['tasmin']['mean'].apply(to_F).plot(ax=ax, label="GDDP Mean Min Temp")
plt.fill_between(gcm_desc['tasmin'].index, 
                 gcm_desc['tasmin']['min'].apply(to_F), 
                 gcm_desc['tasmin']['max'].apply(to_F),
                alpha=0.1)


albany_wu['Max TemperatureF'].plot(ax=ax, label="Weather Ungerground Max for KALB")
albany_wu['Min TemperatureF'].plot(ax=ax, label="Weather Ungerground Min for KALB")

plt.title("Mean Max Temp, Mean Min Temp with Weather Underground historical data")

plt.legend(loc=4)
```

```python
fix, ax = plt.subplots(1,1, figsize=(16., 8.))
pr = gcms.stack(level=0)['pr'].groupby(level=0).describe().unstack()

pr['mean'].plot(label="GDDP Mean across GCMs", ax=ax)
plt.fill_between(pr.index, pr['min'], pr['max'], alpha=0.3)

albany_wu['pr'].plot(label="Weather Underground", ax=ax)

plt.title("Precipitation in kg m-2s-1,  GDDP mean and Actual (Weather Underground Data for KALB), 1997")
plt.legend()
```
